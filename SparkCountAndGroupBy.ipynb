```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.remote("sc://localhost").getOrCreate()
spark
```

```python
df = spark.read.csv("/home/analytics4220/Yujin Kim/Electric_Vehicle_Population_Data.csv", 
                    header=True)
df
```

```python
df.show()
```

```python
# Count each type
counts = df["Electric Vehicle Type (BEV)"].value_counts()
print(counts)
```

```python
bev_count = (df["Electric Vehicle Type"] == "Battery Electric Vehicle (BEV)").sum()
phev_count = (df["Electric Vehicle Type"] == "Plug-in Hybrid Electric Vehicle (PHEV)").sum()

print("BEV count:", bev_count)
print("PHEV count:", phev_count)
```

```python
from pyspark.sql import col, count
```

```python
from pyspark.sql import count
```

```python
from pyspark.sql.functions import col, count
```

```python
bev_count = df.filter(col("Electric Vehicle Type") == "Battery Electric Vehicle (BEV)").count()

phev_count = df.filter(col("Electric Vehicle Type") == "Plug-in Hybrid Electric Vehicle (PHEV)").count()

print("BEV:", bev_count)
print("PHEV:", phev_count)
```

```python
df.groupBy("Electric Vehicle Type").count().show()
```

```python
df.filter(col("Electric Vehicle Type") == "Battery Electric Vehicle (BEV)") \
  .groupBy("Make", "Electric Vehicle Type") \
  .count() \
  .show()
```

```python
df.filter(
    (col("Make") == "TESLA") &
    (col("Electric Vehicle Type") == "Battery Electric Vehicle (BEV)")
).show(1000, False)
```

```python
df.filter(col("Make") == "TESLA") \
  .groupBy("Make", "Electric Vehicle Type") \
  .count() \
  .show()
```

```python
from pyspark.sql.functions import col

df.filter(col("Electric Vehicle Type") == "Battery Electric Vehicle (BEV)") \
  .groupBy("Make") \
  .count() \
  .orderBy("count", ascending=False) \
  .show(1)
```

```python
from pyspark.sql.functions import col

df.filter(col("Electric Vehicle Type") == "Plug-in Hybrid Electric Vehicle (PHEV)") \
  .groupBy("Make") \
  .count() \
  .orderBy("count", ascending=False) \
  .show(1)
```
